```python
# 파이썬 최고 13번 출력해보김
num = 1
while num <= 13:
    print(f"{num} python 최고")
    num += 1
```

```python
score_list =[90,45,70,60,55]

for i in score_list:
    k = score_list.index(i)
    if 60 <= i:
        print(f"{k+1}번 학생은 합격입니다.")
    elif 60 > i:
        print(f"{k+1}번 학생은 불합격입니다.")
```

#### for: range()
- 필요한 만큼의 숫자를 만들어내는 유용한 기능
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
for i in range(len(lunch)): #시작인덱스가 0 과 같다면 생략가능 , 증감값(default : 1) --> 생략가능
    print(i, lunch[i])
```

```python
score_list =[90,45,70,60,55]
for i in range(0,len(score_list),1):
    
    if 60 <= score_list[i]:
        print(f"{i+1}번 학생은 합격입니다.")
    elif 60 > score_list[i]:
        print(f"{i+1}번 학생은 불합격입니다.")
```

```python
for i in range(1,25,1):
    print(i, end=" ")
```

- min,max 함수
- 선언한 자료구조(리스트, 튜플)에서 가장 큰 값, 가장 작은 값을 찾아주는 함수

```python
list2 = [4,5,2,1,99,15,2,7,27]
# list2 에서 가장 작은 수 찾기
min(list2)
```

```python
# 가장 큰 수 구하기
max(list2)

max_num = list2[0]

for i in list2:
    if max_num < i:
        max_num = i
print(max_num)
```

```python
# 1 ~ 100 사이의 숫자중 3의 배수의 합을 구하기
summ =0
for i in range(0,100,3):
    summ +=i
print(summ)
```
