```python
import pandas as pd
```

```python
data = pd.read_csv('data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
```

```python
from matplotlib import font_manager, rc
```

```python
font_list = font_manager.findSystemFonts()
font_list
```

```python
font_name = font_manager.FontProperties(fname = 'c:/windows/Fonts/malgun.ttf').get_name()
font_name
```

```python
rc('font',family = font_name)
```

```python
data.info()
```

```python
car_acc = data[data['사고유형_대분류'] == '차대사람']
car_acc
```

```python
car_acc2 = car_acc[['요일','주야']].groupby(by='요일').value_counts()
```

```python
car_acc2.values
```

```python
y = car_acc2[:][0::2]
y
```

```python
car_acc2 = car_acc[['요일','주야']].groupby(by='요일').sum()
x = car_acc2.index
```

```python
plt.ylim(0,200)
plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.show()
```
