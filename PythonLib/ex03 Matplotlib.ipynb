### 선 그래프 그려보기

```python
# line plot
# 선 그래프 키워드 : plot
y = [2,4,6,8]
plt.plot(y)
plt.show() # 그래프 출력
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.plot(x,y)
plt.show() # 그래프 출력
```

#### 그래프 스타일 지정

```python
# 선 스타일
plt.plot(y,ls='--') # ls : line style
plt.show()
```

```python
# 마커포인트 : marker
plt.plot(y, marker = 'o')
plt.show()
```

```python
# 선 두께 : line width -> lw
plt.plot(y, marker = 'o', lw = 5)
plt.show()
```

```python
# 선 색상 : line color -> c
plt.plot(y,marker = 'o', c = 'g',lw = 5)
plt.show()
```

```python
# marker 의 컬러 지정하기
plt.plot(y, marker = 'o', c= 'g', markerfacecolor = 'm')
plt.show()
```

#### 그래프 옵션 설정

```python
# 그림 범위 지정
plt.plot(x,y,ls =':', c='r',marker = 'D',mfc='b')
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.show()

# 만약 2~8 사이의 값을 보고 싶을때 (zoomin) ==> y축 값의 범위 제한
```

```python
# 틱 설정
plt.plot(x,y,ls =':', c='r',marker = 'D',mfc='b')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
# 눈금선 표시 : grid()
plt.plot(x,y,ls =':', c='r',marker = 'D',mfc='b')
plt.grid()
plt.show()
```

```python
# 여러개의 그래프 출력
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, label ='x')
plt.plot(y, label ='y')
plt.plot(z, label ='z')
plt.legend(loc = 'best') # 범례표시 ' besx ==> 기본값
plt.xlabel('x') # x축 이름
plt.ylabel('y') # y축 이름
plt.title('my graph') # 그래프 제목
plt.show()
```

#### 요일별 사고 건수 count

```python
# 요일 컬럼 접근
# 중복없이 값 확인(월~일) unique()
# value_counts()
data['요일'].unique()
```

```python
y =data['요일'].value_counts()
y
```

```python
y = y[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index # x가 없으므로 요일 정보를 출력하기 위해 인덱스명 추출
plt.bar(x,y)
# 옵션 설정
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 교통사고 건수')
plt.ylim(500,620) # y축 범위 제한
plt.show()
```

#### 차대차 사건중 사상자 수가 많은 발생지 시도를 알아보고 시각화하기
- 차대차사건, 사상자수, 발생지 시도 ==> 접근

- 힌트
    - 사고유형 정보가 있는 컬럼에서 차대차인 행만출력
    - 발생지 시도에 있는 각각의 데이터 회수 확인 = > 확인용
    - '사상자수 발생지 시도 컬럼 인덱싱
    - 발생지 시도별로 묶고 집계함수 연결후 시각화

```python
data.columns
```

```python
car_acc = data[data['사고유형_대분류'] == '차대차']
car_acc
```

```python
car_acc2 = car_acc[['사상자수', '발생지시도']].groupby(by='발생지시도').sum()
car_acc2.index
```

```python
car_acc2.values
```

```python
y = car_acc2['사상자수'].values # 컬럼은 한개지만, 행이 여러개로 2차원 배열 데이터프레임에서 '사상자수' 접근
x
```

```python
x = car_acc2.index # 발생지시도 접근
```

```python
plt.bar(x,y)
plt.title('차대차 교통사고')
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.show()
```
