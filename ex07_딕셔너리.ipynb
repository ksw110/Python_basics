### 딕셔너리란?
- key 와 value를 한 쌍으로 갖는 자료형
- 불변한 key와 가변/불변적인 value로 맵핑된  순서가 없는 집합
- 인덱싱, 슬라이싱이 불가능

### 딕셔너리 선언

```python
# 딕셔너리명 = {key:value, key:value , ...}
a = {} # 빈 딕셔너리
b = {'name':'key'} # 이름정보(문자열)을 담은 딕셔너리
c = {1 : 'key', 2 : 23} # 하나의 딕셔너리에는 다양한 자료형을 담을 수 있다.
d = {'name' : ['인성','성현','탁연']} # value 값으로 리스트를 담을 수 있다.
# 명확하게 뭔가를 정의하고 싶을 때 딕셔너리를 사용
```

```python
print(a)
print(b)
print(c)
print(d)
```

```python
print(type(a))
```

### 딕셔너리 값 추가
- 새로운 키 : 새로운 값

```python
dic_test = {'노래제목' : '사건의 지평선'}
```

```python
dic_test['가수'] = '윤하'
dic_test['날짜'] = '2022.12.01'
```

```python
dic_test
```

```python
# 그냥가져오기
print(dic_test['가수'])
#데이터에 접근할때 사용하는 함수 : get(키)
dic_test.get('노래제목')
```

- 딕셔너리 접근 방법 비교
    - 딕셔너리명[키] = 값이 없을 경우 오류
    - 딕셔너리명.get(키) = 값이 없을경오 none 에러안남
     

```python
gender = aa['gender']
gender
```

```python
gender = aa.get('gender')

print(gender)
```

```python
# 예외처리 try ~ except

try:
    gender = aa['gender']
    gender
except keyerror:
    print('없다')
```

#### 딕셔너리 key, value 모두 가져오기

```python
# items()
aa.items()
```

```python
list(aa.items())
```

```python
for i in aa.items(): # 튜플 형태로 출력을 해줌
    print(i) # 패킹된 형태
```

```python
for key, value in aa.items():
    print(key,value) # 언패킹된 형태로 출력
```

```python
score_dict = {'이름' : ['선영표', '이주희', '김은영', '서대희', '채수민'],
              'Python' : [90, 100, 85, 90, 80],
              'Java' : [85, 100, 80, 95, 85],
              'html/css' : [75, 70, 90, 80, 90]}
```

```python
# 총합을 구하는 함수 : sum()
# 1. python 과목에 접근해서 결과 가져오기
# 2. 1번의 합을 사람 수 만큼 나누기 len()
```

```python
py_avg=sum(score_dict['Python'])/5
ja_avg=sum(score_dict['Java'])/5
hc_avg=sum(score_dict['html/css'])/len(score_dict['html/css'])
```

```python
print('Python 평균 : ', py_avg)
print('Java 평균 : ', ja_avg)
print('html/css 평균 : ', hc_avg)
```

```python
# 각 과목의 평균을 for 문을 활용하여 구해보기
avg = 0
for i in score_dict.keys():
    print(i)
```

```python
# 이름은 빼고 과목만 가져오기위해 인덱싱, 슬라이싱을 위해 리스트로 형변환
avg = 0
for i in list(score_dict.keys())[1:]:
    avg = sum(score_dict[i])/len(score_dict[i])
    print(f'{i}평균 : {avg}')
```
